```python
# Cell 2: Helper Functions
def test_endpoint(name: str, method: str, url: str, **kwargs):
    """Test an API endpoint and print results"""
    print(f"\n{'='*60}")
    print(f"ðŸ§ª Testing: {name}")
    print(f"{'='*60}")
    print(f"Method: {method}")
    print(f"URL: {url}")
    
    try:
        if method == "GET":
            response = requests.get(url, **kwargs)
        elif method == "POST":
            response = requests.post(url, **kwargs)
        elif method == "PUT":
            response = requests.put(url, **kwargs)
        elif method == "DELETE":
            response = requests.delete(url, **kwargs)
        
        print(f"Status: {response.status_code}")
        
        if response.status_code < 400:
            print("âœ… SUCCESS")
            print(f"Response: {json.dumps(response.json(), indent=2)[:500]}...")
            return response.json()
        else:
            print("âŒ FAILED")
            print(f"Error: {response.text[:200]}")
            return None
            
    except Exception as e:
        print(f"âŒ EXCEPTION: {str(e)}")
        return None


def create_test_audio_file():
    """Create a dummy audio file for testing"""
    # Create a minimal valid MP3 file (silence)
    import io
    audio_data = b'\xff\xfb\x90\x00' + b'\x00' * 100
    return io.BytesIO(audio_data)

```

```python
import base64, json

def jwt_sub(jwt: str) -> str:
    # returns the user UUID inside the JWT (the "sub" claim)
    payload_b64 = jwt.split('.')[1]
    payload_b64 += '=' * (-len(payload_b64) % 4)  # base64url padding
    payload = json.loads(base64.urlsafe_b64decode(payload_b64))
    return payload["sub"]
```

```python
def quick_test_all():
    """Run all tests quickly"""
    print("ðŸš€ Running Quick Test Suite...\n")
    
    tests = [
        ("Health Check", "GET", f"{API_BASE_URL}/health"),
        ("Dashboard Summary", "GET", f"{API_BASE_URL}/api/v1/dashboard/summary", headers_user),
        ("Unified Stats", "GET", f"{API_BASE_URL}/api/v1/unified/stats", headers_user),
    ]
    
    results = {"passed": 0, "failed": 0}
    
    for name, method, url, *args in tests:
        headers = args[0] if args else {}
        try:
            if method == "GET":
                response = requests.get(url, headers=headers, timeout=5)
            if response.status_code < 400:
                print(f"âœ… {name}")
                results["passed"] += 1
            else:
                print(f"âŒ {name} - Status: {response.status_code}")
                results["failed"] += 1
        except Exception as e:
            print(f"âŒ {name} - Error: {str(e)[:50]}")
            results["failed"] += 1
    
    print(f"\nðŸ“Š Results: {results['passed']} passed, {results['failed']} failed")
    return results

# Uncomment to run quick test
quick_test_all()

```
